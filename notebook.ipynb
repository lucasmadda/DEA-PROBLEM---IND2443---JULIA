```julia
using JuMP, GLPK, XLSX
```

## Data Import

```julia
xf = XLSX.readxlsx("data.xlsx");
```

## Sets

```julia
DMU_d = xf["Planilha1"]["A2:A31"];

input_i = xf["Planilha1"]["B1:C1"];

output_j = xf["Planilha1"]["D1:H1"];
```

## Parameters

```julia
a = Dict([input_i[i],DMU_d[d]] =>
    transpose(xf["Planilha1"]["B2:C31"])[i,d] for i in 1:length(input_i) for d in 1:length(DMU_d));

b = Dict([output_j[j],DMU_d[d]] =>
    transpose(xf["Planilha1"]["D2:H31"])[j,d] for j in 1:length(output_j) for d in 1:length(DMU_d));

p = DMU_d;
```

## Model

```julia
results = Dict()

for p_ in p

    m = Model(GLPK.Optimizer);

    @variable(m,
        x[input_i,DMU_d] >= 0);
    
    @variable(m,
        y[output_j,DMU_d] >= 0);
    
    
    @constraint(m,
    [d in DMU_d],
    sum(b[[j,d]] * y[j,p_] for j in output_j) - sum(a[[i,d]] * x[i,p_] for i in input_i) <= 0);
    
    @constraint(m,
        sum(a[[i,p_]] * x[i,p_] for i in input_i) == 1)
    
    
    @objective(m,
        Max,
        sum(b[[j,p_]] * y[j,p_] for j in output_j))
    
    
    optimize!(m)
    
    results[p_] = objective_value(m)
    
end
```

```julia
for p_ in p
    println(results[p_])
end
```
